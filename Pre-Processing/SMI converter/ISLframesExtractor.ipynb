```python
# FRAME SPATIAL EMBEDDING ############################

from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.applications.mobilenet import preprocess_input
import cv2
import numpy as np

# Load pretrained MobileNet model (exclude classification layers)
base_model = MobileNet(weights='imagenet', include_top=False)

# Add GlobalAveragePooling2D layer to convert feature maps to a fixed-size vector
x = base_model.output
x = GlobalAveragePooling2D()(x)

# Create the feature extractor model
feature_extractor = Model(inputs=base_model.input, outputs=x)

# Load and preprocess an image
image_path = 'D:/For Rahul G/output_frames/2/2_2.00_14.00/frame_0079.jpg'
image = cv2.imread(image_path)

# Get image dimensions
height, width, channels = image.shape

# Display image dimensions
print(f"Height: {height} pixels")
print(f"Width: {width} pixels")
print(f"Channels: {channels} channels")


# Define the coordinates of the ROI (top-left and bottom-right)
x1, y1 = 535, 0  # Top-left corner of ROI
x2, y2 = 1385, 1080  # Bottom-right corner of ROI
#210*260
# Crop the image to the specified ROI
image = image[y1:y2, x1:x2]
image = cv2.resize(image, (224, 224))  # Resize to the input size of MobileNet
# Display the image in a window
cv2.imshow('Image', image)


image = preprocess_input(image)

# Expand dimensions to match the expected input shape (batch size of 1)
image = np.expand_dims(image, axis=0)

# Extract spatial embedding
spatial_embedding = feature_extractor.predict(image)
print("******************:  ",spatial_embedding)
#cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
###### TO SEPARATE FRAMES IN DIFFERENT FOLDERS FOR MULTIPLE VIDEOS PRESENT IN video_files and create their feature files ######################
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.applications.mobilenet import preprocess_input
import cv2
import numpy as np
import torch
def tokenize_features(features):
        ft_list = np.split(features, 1)
        return [ft.squeeze() for ft in ft_list]
# Load pretrained MobileNet model (exclude classification layers)

base_model = MobileNet(weights='imagenet', include_top=False)

# Add GlobalAveragePooling2D layer to convert feature maps to a fixed-size vector
x = base_model.output
x = GlobalAveragePooling2D()(x)

# Create the feature extractor model
feature_extractor = Model(inputs=base_model.input, outputs=x)

import re
import os
import cv2
def convert_time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(":"))
    total_seconds = (minutes * 60) + seconds
    return total_seconds

def extract_pattern_from_file(file_path, pattern):
    collected_patterns = []
    with open(file_path, 'r') as file:
        content = file.read()
        
        # Find all occurrences of the pattern in the content
        matches = re.findall(pattern, content)
        
        # Append the matched patterns to the collected_patterns list
        collected_patterns.extend(matches)
        
    return collected_patterns

def extract_frames(input_video, output_folder, start_time, end_time, indx,frame_rate=25):
    # Open the video file
    video_capture = cv2.VideoCapture(input_video)

    # Get frames per second (FPS) and total number of frames in the video
    fps = int(video_capture.get(cv2.CAP_PROP_FPS))
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))

    # Calculate frame indices for start and end times
    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)

    if end_frame >= total_frames:
        print("End time exceeds video duration.")
        return

    # Set the capture position to the start frame
    video_capture.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
    subfolder_name = f"{output_folder}/{indx}_{start_time:.2f}_{end_time:.2f}"
    os.makedirs(subfolder_name, exist_ok=True)
    # Loop through frames and save them as images
    for frame_idx in range(start_frame, end_frame):
        ret, frame = video_capture.read()
        if not ret:
            break
       # To collect reduced number of frames %2
        if frame_idx%2: 
            frame_filename = f"{subfolder_name}/frame_{frame_idx:04d}.jpg"
            cv2.imwrite(frame_filename, frame)
            image = cv2.imread(frame_filename)

            # Get image dimensions
            height, width, channels = image.shape

            # Display image dimensions
            """print(f"Height: {height} pixels")
            print(f"Width: {width} pixels")
            print(f"Channels: {channels} channels")"""


            # Define the coordinates of the ROI (top-left and bottom-right)
            x1, y1 = 535, 0  # Top-left corner of ROI
            x2, y2 = 1385, 1080  # Bottom-right corner of ROI
            #210*260
            # Crop the image to the specified ROI
            image = image[y1:y2, x1:x2]
            image = cv2.resize(image, (224, 224))  # Resize to the input size of MobileNet
            # Display the image in a window
            #cv2.imshow('Image', image)


            image = preprocess_input(image)

            # Expand dimensions to match the expected input shape (batch size of 1)
            image = np.expand_dims(image, axis=0)

            # Extract spatial embedding
            spatial_embedding = feature_extractor.predict(image)
            # Save the features to a NumPy file
            
            feature_filename = f"{subfolder_name}/feature_{frame_idx:04d}.npy"        
            #list of features can be created here#########################################################################
            np.save(feature_filename, spatial_embedding)
    video_capture.release()
    
if __name__ == "__main__":
   
    pattern_to_extract = r'\b\w+:\w+\w+\b'  # Example pattern to extract 1:02
    input_folder = "D:/For Rahul G/"
    output_folder = "D:/For Rahul G/output_frames/"
    
    video_files = [file for file in os.listdir(input_folder) if file.endswith(".mp4")]
    file_paths = [file for file in os.listdir(input_folder) if file.endswith(".txt")]
    print(video_files)
    print(file_paths)
    for indx, video_file in enumerate(video_files):
        input_video ="D:/For Rahul G/"+ video_file
        file_path = "D:/For Rahul G/" + file_paths[indx]
        extracted_patterns = extract_pattern_from_file(file_path, pattern_to_extract)
        collected_time = []
        for pattern in extracted_patterns:
            input_time=pattern
            seconds = convert_time_to_seconds(input_time)
            collected_time.append(seconds)    
        for index,pattern in enumerate(extracted_patterns):
            if index != len(extracted_patterns)-1:
                start_time = collected_time[index]  # Start time in seconds
                end_time = collected_time[index+1]    # End time in seconds
            extract_frames(input_video, output_folder+str(indx), start_time, end_time,indx)
        print(collected_time)
        print("Frames extracted successfully.")
        indx=indx+1
```

```python
#########################TO COPY TEXT(single file) IN EXCEL SHEET IN SEPARATE CELLS differnt columns(with names as per frame folder) #######################
import openpyxl
import re
import os
c=0
def convert_time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(":"))
    total_seconds = (minutes * 60) + seconds
    return total_seconds

def copy_text_with_pattern_to_excel(txt_file_path, xlsx_file_path, pattern):
    global c
    # Read text from the .txt file
    with open(txt_file_path, 'r') as txt_file:
        text_content = txt_file.read()

    # Split the text content using the specified pattern as delimiter
    split_content = re.split(pattern, text_content)
    print(len(split_content))     
    # Create a new Excel workbook and sheet
    workbook = openpyxl.Workbook()
    sheet = workbook.active
  
    matches = re.findall(pattern, text_content)       
        # Print the found patterns
    collected_time = []
    for pattern in matches:
            input_time=pattern
            seconds = convert_time_to_seconds(input_time)
            collected_time.append(seconds) 
    
    
    """for match in matches:
            print(convert_time_to_seconds(match))
            print(match)"""
    print(len(matches))     
   
    for row_idx, content_part in enumerate(matches, start=1):     
        sheet.cell(row=row_idx+1, column=1, value=str(1)+"_"+str(convert_time_to_seconds(content_part))+".00")
    
    
    # Write the split content to the Excel sheet
    for row_idx, content_part in enumerate(split_content, start=1):
        sheet.cell(row=row_idx, column=2, value=content_part)

    # Save the Excel workbook
    workbook.save(xlsx_file_path+"output_excel"+str(c)+".xlsx")
    print(f"Text copied to {xlsx_file_path} successfully.")
    c=c+1

if __name__ == "__main__":
    input_folder = "D:/For Rahul G/"
    output_folder = "G:/Project for own dataset 1/Projects/slt-master/data/output_excels/"
    txt_files = [file for file in os.listdir(input_folder) if file.endswith(".txt")]
    print(txt_files)
    for indx, txt_file in enumerate(txt_files):
        txt_file_path ="D:/For Rahul G/"+ txt_file
    
       # txt_file_path = "D:/For Rahul G/script01.txt"  
        xlsx_file_path = "G:/Project for own dataset 1/Projects/slt-master/data/output_excels/"
        pattern = r'\b\w+:\w+\w+\b'  # Example pattern (semicolon surrounded by optional spaces)
        copy_text_with_pattern_to_excel(txt_file_path, xlsx_file_path, pattern)
    
    
```
